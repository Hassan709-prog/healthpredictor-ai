```python
# ==========================================================
# Step 1: Import Required Libraries
# ==========================================================

import pandas as pd
import numpy as np

# Data Preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score

# Machine Learning Models
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

# Model Evaluation
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix
)

print("All libraries imported successfully!")
```

```python
# ==========================================================
# Step 3: Dataset Shape
# ==========================================================

print("Dataset Shape:", dataset.shape)
```

```python
# Load Dataset
dataset = pd.read_csv("dataset.csv")

print("Dataset loaded successfully!")
```

```python
print("Dataset Shape:", dataset.shape)
```

```python
# ==========================================================
# Step 4: Display Dataset Information
# ==========================================================

print("=" * 60)
print("DATASET INFORMATION")
print("=" * 60)

dataset.info()
```

```python
# ==========================================================
# Step 6: Basic Dataset Statistics
# ==========================================================

print("=" * 60)
print("BASIC DATASET STATISTICS")
print("=" * 60)

print("Dataset Shape          :", dataset.shape)
print("Total Records          :", len(dataset))
print("Total Columns          :", len(dataset.columns))
print("Total Diseases         :", dataset["Disease"].nunique())
```

```python
# ==========================================================
# Step 7: Count Total Unique Symptoms
# ==========================================================

all_symptoms = set()

for column in dataset.columns[1:]:
    all_symptoms.update(dataset[column].dropna().unique())

print("=" * 60)
print("SYMPTOM ANALYSIS")
print("=" * 60)

print("Total Unique Symptoms :", len(all_symptoms))
```

```python
# ==========================================================
# Step 8: Missing Values Analysis
# ==========================================================

print("=" * 60)
print("MISSING VALUES")
print("=" * 60)

print(dataset.isnull().sum())
```

```python
# ==========================================================
# Step 10: Replace Missing Values
# ==========================================================

dataset.fillna("", inplace=True)

print("Missing values replaced successfully!")
```

```python
# ==========================================================
# Step 11: Verify Missing Values
# ==========================================================

print("Remaining Missing Values:")
print(dataset.isnull().sum().sum())
```

```python
# ==========================================================
# Step 12: Create Master Symptom List
# ==========================================================

all_symptoms = set()

for column in dataset.columns[1:]:
    all_symptoms.update(dataset[column].unique())

# Remove empty string
all_symptoms.discard("")

# Convert to sorted list
all_symptoms = sorted(list(all_symptoms))

print("Total Symptoms:", len(all_symptoms))
```

```python
# ==========================================================
# Step 13: Create Binary Feature Matrix
# ==========================================================

X = []

for _, row in dataset.iterrows():

    symptoms = set(row[1:])

    feature_vector = []

    for symptom in all_symptoms:

        if symptom in symptoms:
            feature_vector.append(1)
        else:
            feature_vector.append(0)

    X.append(feature_vector)

print("Feature Matrix Created Successfully!")
```

```python
# ==========================================================
# Step 14: Convert Feature Matrix into DataFrame
# ==========================================================

X = pd.DataFrame(X, columns=all_symptoms)

print("Feature Matrix Shape:", X.shape)
```

```python
# ==========================================================
# Step 15: Prepare Target Variable
# ==========================================================

y = dataset["Disease"]

print("Target Variable Created Successfully!")
print("Total Classes:", y.nunique())
```

```python
# ==========================================================
# Step 16: Encode Disease Labels
# ==========================================================

label_encoder = LabelEncoder()

y = label_encoder.fit_transform(y)

print("Disease Labels Encoded Successfully!")
```

```python
# ==========================================================
# Step 17: Verify Final Dataset
# ==========================================================

print("Features Shape :", X.shape)
print("Labels Shape   :", y.shape)
```

```python
# ==========================================================
# Step 18: Split Dataset into Training and Testing Sets
# ==========================================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Dataset Split Successfully!")
print("Training Samples :", X_train.shape[0])
print("Testing Samples  :", X_test.shape[0])
```

```python
# ==========================================================
# Step 19: Train Decision Tree Classifier
# ==========================================================

dt_model = DecisionTreeClassifier(random_state=42)

dt_model.fit(X_train, y_train)

print("Decision Tree trained successfully!")
```

```python
# ==========================================================
# Step 20: Evaluate Decision Tree
# ==========================================================

dt_predictions = dt_model.predict(X_test)

dt_accuracy = accuracy_score(y_test, dt_predictions)

print("Decision Tree Accuracy:", round(dt_accuracy * 100, 2), "%")
```

```python
# ==========================================================
# Step 21: Train Random Forest Classifier
# ==========================================================

rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf_model.fit(X_train, y_train)

print("Random Forest trained successfully!")
```

```python
# ==========================================================
# Step 22: Evaluate Random Forest
# ==========================================================

rf_predictions = rf_model.predict(X_test)

rf_accuracy = accuracy_score(y_test, rf_predictions)

print("Random Forest Accuracy:", round(rf_accuracy * 100, 2), "%")
```

```python
# ==========================================================
# Step 23: Train XGBoost Classifier
# ==========================================================

xgb_model = xgb_model = XGBClassifier(
    objective="multi:softprob",
    num_class=41,
    random_state=42,
    eval_metric="mlogloss"
)


xgb_model.fit(X_train, y_train)

print("XGBoost trained successfully!")
```

```python
print(xgb_model.get_params()["objective"])
```

```python
# ==========================================================
# Step 24: Evaluate XGBoost
# ==========================================================

xgb_predictions = xgb_model.predict(X_test)

xgb_accuracy = accuracy_score(y_test, xgb_predictions)

print("XGBoost Accuracy:", round(xgb_accuracy * 100, 2), "%")
```

```python
# ==========================================================
# Step 25: Compare Model Accuracy
# ==========================================================

comparison = pd.DataFrame({
    "Model": [
        "Decision Tree",
        "Random Forest",
        "XGBoost"
    ],
    "Accuracy (%)": [
        round(dt_accuracy * 100, 2),
        round(rf_accuracy * 100, 2),
        round(xgb_accuracy * 100, 2)
    ]
})

comparison
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
# ==========================================================
# Step 26: 10-Fold Cross Validation
# ==========================================================

kfold = StratifiedKFold(
    n_splits=10,
    shuffle=True,
    random_state=42
)

scores = cross_val_score(
    xgb_model,
    X,
    y,
    cv=kfold,
    scoring="accuracy"
)

print("Cross Validation Scores:")
print(scores)

print("\nAverage Accuracy:", scores.mean()*100)
print("Standard Deviation:", scores.std()*100)
```

```python
# ==========================================================
# Step 27: Classification Report
# ==========================================================

print(classification_report(y_test, xgb_predictions,
                            target_names=label_encoder.classes_))
```

```python
# ==========================================================
# Step 28: Confusion Matrix
# ==========================================================

cm = confusion_matrix(y_test, xgb_predictions)

print("Confusion Matrix Shape:", cm.shape)
```

```python
# ==========================================================
# Step 29: Top 20 Important Symptoms
# ==========================================================

importance = pd.DataFrame({
    "Symptom": X.columns,
    "Importance": xgb_model.feature_importances_
})

importance = importance.sort_values(
    by="Importance",
    ascending=False
)

importance.head(20)
```
